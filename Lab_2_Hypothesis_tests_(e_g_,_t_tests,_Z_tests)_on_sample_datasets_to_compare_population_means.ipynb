```python
import numpy as np
import scipy.stats as stats
```

# One sample t-test

```python
scores = [78, 72, 74, 76, 73, 79, 77, 71, 70, 74]
```

```python
population_mean = 75
n = len(scores)
```

```python
sample_mean = np.mean(scores)
sample_std = np.std(scores, ddof=1)
```

```python
# Calculate t-statistic
t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
```

```python
# Degrees of freedom
df = n - 1
```

```python
# Two-tailed p-value
p_value = 2 * stats.t.sf(abs(t_stat), df) # Results
```

```python
print(f"Sample Mean: {sample_mean:.2f}")
print(f"tstatistic: {t_stat:.2f}")
print(f"p-value: {p_value:.4f}")
```

```python
if p_value < 0.05:
  print("Reject the null hypothesis: The sample mean is significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

# Two sample t-test

```python
 # Data
class_a = [85, 88, 90, 92, 86, 84, 87, 89, 91, 93]
class_b = [78, 80, 82, 79, 81, 77, 83, 76, 75, 80]
```

```python
# Two-sample t-test
t_stat, p_value = stats.ttest_ind(class_a, class_b, equal_var=False)
```

```python
# Results
print(f"t-statistic:{t_stat:.2f}")
print(f"pvalue: {p_value:.4f}")
if p_value < 0.05:
  print("Reject the null hypothesis: The means are significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

# One-Sample z-Test
A factory claims the average weight of a product is 500g. A sample of 40 products
has a mean weight of 495g, and the known population standard deviation is 10g.
Test the claim (α=0.05).

```python
# Data
sample_mean = 495
population_mean = 500
population_std = 10
n = 40
```

```python
# Calculate z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
# Calculate p-value (two-tailed)
p_value = 2 * stats.norm.sf(abs(z_stat))
```

```python
# Results
print(f"z-statistic: {z_stat:.2f}")
print(f"pvalue: {p_value:.4f}")
if p_value < 0.05:
  print("Reject the null hypothesis: The sample mean is significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

# Two-Sample z-Test
Scenario:
You compare the average weights of products from two factories to see if there is
a significant difference.

```python
 # Data
mean1, std1, n1 = 495, 10, 50
mean2, std2, n2 = 500, 12, 60
```

```python
# Calculate z-statistic
z_stat = (mean1 - mean2) / ((std1**2 / n1 + std2**2 / n2) ** 0.5)
# Two-tailed p-value
p_value = 2 * stats.norm.sf(abs(z_stat))
# Results
print(f"z-statistic: {z_stat:.2f}")
print(f"pvalue: {p_value:.4f}")
if p_value < 0.05:
  print("Reject the null hypothesis: The means are significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```
